```python
import os
import numpy as np
import scipy as sp
import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python
path = r'./3-class-clean'
```

```python
files = os.listdir(path)
```

```python
files[:9]
```

```python
class_map = {
    "yukari":"u",
    "asagi":"d",
    "sag":"r",
    "sol":"l",
    "kirp":"b"
}
```

```python
test = np.loadtxt(path+ r"/" + files[0], dtype=int)
```

```python
data = pd.DataFrame()
ids = []
horizontal = []
vertical = []
labels = []
```

```python
for i in range(0,len(files),2):
    file1 = path+ r"/" + files[i]
    file2 = path+ r"/" + files[i+1]
    sig_id = re.findall(r'\d+', files[i])
    ids.append(int(sig_id[0]))
    h = np.loadtxt(file1, dtype=int)
    v = np.loadtxt(file2, dtype=int)
    h -= int(np.average(h))
    v -= int(np.average(v))
    horizontal.append(h)
    vertical.append(v)
    labels.append(class_map[files[i].split(sig_id[0])[0]])
```

```python
data["ID"] = ids
data["Horizontal"] = horizontal
data["Vertical"] = vertical
data["Label"] = labels
```

```python
t = np.arange(data.iloc[0]["Horizontal"].size)
```

```python
for i in range(0,data.shape[0],20):
    plt.figure(figsize = (12,6))
    plt.plot(t,data.iloc[i]["Horizontal"],'r')
    plt.xlabel("time")
    plt.ylabel(data.iloc[i]["Label"])
    plt.tick_params(axis='x', colors='white')
    plt.tick_params(axis='y', colors='white')
    plt.figure(figsize = (12,6))
    plt.plot(t,data.iloc[i]["Vertical"],'r')
    plt.xlabel("time")
    plt.ylabel(data.iloc[i]["Label"])
    plt.tick_params(axis='x', colors='white')
    plt.tick_params(axis='y', colors='white')
```
